```python
#  Sales Data Analysis with Pandas

import pandas as pd
import matplotlib.pyplot as plt

# 1. Load CSV
df = pd.read_csv("sales.csv")
print("First 5 rows:")
display(df.head())

# 2. Basic info
print("Shape of dataset:", df.shape)
print("Columns:", df.columns)

# 3. Total sales per category
category_sales = df.groupby("Category")["Sales"].sum()
print("\nSales by Category:")
display(category_sales)

# Plot sales by category
category_sales.plot(kind="bar", title="Total Sales by Category", xlabel="Category", ylabel="Sales", figsize=(6,4))
plt.show()

# 4. Total quantity sold per product
product_quantity = df.groupby("Product")["Quantity"].sum()
print("\nQuantity sold per Product:")
display(product_quantity)

# Plot quantities
product_quantity.plot(kind="barh", title="Quantity Sold per Product", xlabel="Quantity", ylabel="Product", figsize=(6,4))
plt.show()

# 5. Sales trend over time
daily_sales = df.groupby("Date")["Sales"].sum()
print("\nDaily Sales:")
display(daily_sales)

# Plot sales trend
daily_sales.plot(kind="line", marker="o", title="Daily Sales Trend", xlabel="Date", ylabel="Sales", figsize=(8,4))
plt.xticks(rotation=45)
plt.show()
```
